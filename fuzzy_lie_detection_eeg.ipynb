## Important definitions
* The P300 (P3) wave is an event-related potential (ERP) component elicited in the process of decision making. It is considered to be an endogenous potential, as its occurrence links not to the physical attributes of a stimulus, but to a person's reaction to it.

* Event-related potentials (ERPs) are very small voltages generated in the brain structures in response to specific sensory, cognitive, or motor events.
 They are measured by means of electroencephalography (EEG) and can be used to identify and classify perceptual, memory, and linguistic operations. ERPs arise from the synchronous activities of neuronal populations engaged in information processing.
They provide a safe and noninvasive approach to study psychophysiological correlates of mental processes. ERPs can be elicited by a wide variety of sensory, cognitive, or motor events.
The magnetoencephalography equivalent of ERP is the ERF, or event-related field.

* The Concealed Information Test (CIT) is a psychophysiological technique for examining whether a person has knowledge of crime-relevant information. Many laboratory studies have shown that the CIT has good scientific validity. However, the CIT has seldom been used for actual criminal investigations.

* Fuzzy Logic - Fuzzy logic is a form of many-valued logic where the truth value of variables can be any real number between 0 and 1, allowing for multiple possible truth values to be processed through the same variable. It is used to handle the concept of partial truth, where the truth value may range between completely true and completely false.
Fuzzy logic is an approach to variable processing that allows for multiple possible truth values to be processed through the same variable. Fuzzy logic attempts to solve problems with an open, imprecise spectrum of data and heuristics that makes it possible to obtain an array of accurate conclusions.
Fuzzy logic is designed to solve problems by considering all available information and making the best possible decision given the input.

* Common Spatial Pattern (CSP) is a signal processing technique, which transforms the EEG data into various subcomponents using linear transformation. It projects multichannel EEG data into lower dimensional subspace. It does so by maximizing the differences in variance of two classes.

The elicitation of P300 is important in "Lie Detection" applications. Such applications utilize P300 component of ERP for analyzing the deceit behavior of a person.

## EEG-Based Concealed Information Tests
>EEG-based CITs are conducted by presenting a series of stimuli to the subject. These stimuli are basically classified into three types: probe, irrelevant, and target. ***The stimuli having the concealed information and being known to the guilty are referred to as probe.***The probe stimuli work as irrelevant stimuli when the subject is innocent. The target stimuli refer to the stimuli being well known to both guilty and innocent subjects. The probe stimuli having been known by the guilty naturally will lead to the guilty concealing that information. This in turn evokes a P300 response in the guilty subjectâ€™s brain, the reason being the presence of that concealed information in his/her memory. These P300 responses can be recorded and identified using any EEG device, whereas for innocent, these probe stimuli behave as irrelevant stimuli and no such response is generated. In addition to that, some irrelevant stimuli are presented to guilty to create a difference between probe and irrelevant responses. Target stimuli are certain fixed stimuli, which generates a P300 response in both guilty and innocent subjects. For guilty subjects, probe and target responses are the same and vice versa for innocent subjects. These differences are utilized to obtain a classification between the concealed behavior and the innocent behavior.

## Methodology
During the session, subjects are advised to avoid muscle movement and it has been taken care that the experimental room is isolated to avoid outside noises.

The experiment to record EEG data involves ten subjects who are supposed to perform the lie detection test for two sessions consisting of 30 trials each. Three types of stimuli, i.e., probe, irrelevant, and target, are given to all ten subjects for two sessions. The first session is a truth session and the second session is a lie session.

For both the sessionss all ten subjects have voluntarily participated. In the first truth session, the set of images has been presented as stimuli and subjects are instructed not to perform any kind of covert behavior and reply correctly whether they know the person whose face is presented in the image. During the second lie session, subjects have been instructed to perform any kind of covert activities and are informed to lie for the images of Fig. 2. Experimental trial structure. person whom they know also.


The EEG data recorded are converted into numerical attributes for further analysis. It is then filtered using a bandpass filter in the range of 0.5â€“30 Hz. The filter is applied at this range, as this is the frequency that mostly appears for mental tasks. The preprocessed EEG data are processed by applying CSP as a feature extraction approach. After converting multichannel EEG data into different components using CSP, the final feature set is given to the proposed fuzzy integrator system.
